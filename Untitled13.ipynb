```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import fredapi as fd
import plotly.express as px
```

```python
fred = fd.Fred(api_key='')
```

```python
data=fred.search('Consumer Price Index for All Urban Consumers: All Items in U.S. City Average')
data.head(10)

```

```python
data['title'][0]
```

```python
cpi=fred.get_series('CPIAUCSL')
cpi.names='values'
cpi
```

```python
df=pd.DataFrame(cpi).reset_index()
df
```

```python
df2=df[df['index']>'1989-12-01']
df2
```

```python
fig = px.line(df2,x='index' ,y= 0, title='CPI Avg. 1990-2022')
fig.show()
```

```python
df2=df2.rename(columns={'index':'ds' , 0 :'y'})
df2
```

```python
from prophet import Prophet
m = Prophet()
m.fit(df2)
```

```python
future=m.make_future_dataframe(periods=10, freq='ms')#weekly w
```

```python
future.tail()
```

```python
result=m.predict(future)
result[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
fig=m.plot(result)
```

```python
fig2=m.plot_components(result)
```

```python
from prophet.diagnostics import cross_validation,performance_metrics
cv_result=cross_validation(model=m, initial=pd.to_timedelta(30*20 ,unit='D'),period=pd.to_timedelta(30*5 ,unit='D'),horizon=pd.to_timedelta(30*12 ,unit='D'))
df_p=performance_metrics(cv_result)
```

```python
df_p
```

```python
from prophet.plot import plot_cross_validation_metric
fig=plot_cross_validation_metric(cv_result, metric='mae')
```
